## Train a Default Network

```python
# create the environment
env = BananaEnv("../../deep-reinforcement-learning/p1_navigation/Banana_Windows_x86_64/Banana.exe", True)
```

```python
params = {
    "episodes": 100,
    "batch_size": 64,
    "buffer_size": 10000,
    "learning_rate": 5e-4,
    "discount_rate": 0.99,
    "epsilon_decay": 0.995,
    "epsilon_decay_rate": 1,
    "min_epsilon": 1e-2,
    "update_rate": 4
}
```

```python
agent = DqnAgent(37, 4, params)
```

```python
scores, epsilons = agent.train(env)
```
